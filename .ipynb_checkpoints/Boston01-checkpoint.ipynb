## Missing Attributes

In order to deal with missing attributes you have 3 options:
1) Remove all missing datapoints
2) Delete the whole attribute
3) Replace the missing attribute with some value (0/mean/median)

```python
a = housing.dropna(subset=["RM"]) #Option 1
a.shape
```

```python
b = housing.drop("RM", axis=1)  #Option 2
b.shape
```

```python
median = housing["RM"].median()  #Option3
housing["RM"].fillna(median)
housing.shape
```

## SciKit-learn Design

Primarily, there are 3 types of Objects:
1) Estimators : It estimates some parameters based on some dataset. Eg. Imputer. It has a fit method and transform method. Fit method-fits the dataset and calculates internal parameters.

2) Transformers: transform method takes the input and returns output based on the learnings from the fit() method. It has a convenience function called fit_transform() which fits and then transforms.

3) Predictors: LinearRegression model is an example of a predictor. fit() and predictor() are two common functions. It also gives score() function which will evaluate the predictions.

# Feature Scaling

1) Min-Max Scaling (Normalization): (Value-Min)/(Max-Min). sklearn has MinMaxScalar for this.
2) Standardization: (Value-mean)/stad deviation. sklearn has StandardScalar for this.
