## Pandas Tutorial (Harry)

```python
import numpy as np
import pandas as pd
```

```python
dict1 = {
    "Name":['Parth','Patel','Anna','Nayan','Elon'],
    "Marks":[50,20,90,100,30],
    "City":['Rampur','Bareily','Ahmedabad','Baroda','Surat']
}                             ### Creates a Dictionary dict1
df = pd.DataFrame(dict1)      ### Converts this dictionary to DataFrame
print(df)                     ### Prints this DataFrame
```

```python
df.to_csv("data2.csv", index=False)   ### Stores this DataFrame into .csv file
```

```python
df.head(2)     ###  Shows top 2 rows stored in DataFrame
```

```python
df.tail(2)     ###  Shows bottom 2 rows stored in DataFrame
```

```python
df.describe()    ### Shows all the statistical values related to this DataFrame
```

```python
parth = pd.read_csv("data2.csv")    ###  Reads .csv file and stores values into another dataframe and then prints it
print(parth)
```

```python
parth["Marks"][0] = 100    ### Changes one element in previous DataFrame and prints updated DataFrame
print (parth)
```

```python
parth.to_csv("data2.csv", index=False)    ### Stores the DataFrame into .csv file without indices
print(pd.read_csv("data2.csv"))
```

```python
parth.index = ['First', 'Second', 'Third', 'Fourth', 'Fifth']   ### Assigns indices name and prints it
print(parth)
```

# Series: Only one row or one column, like vector
# Dataframe: More than one row and column, like spreadsheet (n*n)

```python
ser = pd.Series(np.random.rand(34))     ### Generates the series of random integers from 0 to 33
print(ser)
print(type(ser))      ### Prints the type of series
```

```python
newdf = pd.DataFrame(np.random.rand(334,5))     ### Generates DataFrame of random number of 334 rows and 5 columns
print(newdf)
```

```python
newdf.describe()
```

```python
newdf.head()
```

```python
newdf.tail()
```

```python
newdf.dtypes         ### Shows the datatype of the DataFrame
```

```python
newdf.to_numpy()    ### Converts DataFrame into Numpy array
```

```python
newdf.T        ###  Shows the Transpose of the DataFrame
```

```python
# Sorts the DataFrame wrt to row/column, in ascending/descending order
newdf.sort_index(axis=0, ascending=False)   # 0 = row and 1 = column
```

```python
newdf2 = newdf.copy()   ### Creates the copy of the array
print(newdf2)
```

```python
newdf2[0][0] = 0.80     ### Changes element value in copied DataFrame but doesn't change the original DataFrame
print(newdf2)
print(newdf)
```

```python
newdf2.columns = list('A BCDE')        ### Renames the columns of the DataFrame
print(newdf2)
```

```python
newdf2.drop('E', axis=1)     ### Drops 1 column from the DataFrame (must give wrt to which axis)
### 0=Row & 1=Column
```

```python
newdf2.drop([1,5],axis=0)     ### Drops 2 Rows from the DataFrame (must give wrt to which axis)
### 0=Row & 1=Column
```

```python
print(newdf2)
```

```python
newdf2.loc[333,'E']  ### .loc function used to access particular element in the DataFrame
```

```python
newdf2.loc[[1,3],['C','E']]   ### .loc function used to access particular elements in the DataFrame
```

```python
newdf2.iloc[[1,3],[2,4]]    ### .iloc function requires indices of the elements required as I/P
```

```python
# newdf2['E'] = None     ### Assigns None to all the elements of the column E
print(newdf2)
```

```python
dict2 = {
    'A': [1000, 200, 32], 'B': [45, 56, 20]
}
newdf3 = pd.DataFrame(dict2)
print (newdf3)
```

```python
print(newdf3.describe())
```

```python
print(newdf3.mean())     ### Prints the mean value of each column of the DataFrame
```

```python
print(newdf3.corr())      ### Prints the correlation of each column of the DataFrame wrt eachother
```

```python
print(newdf3.count())     ### Prints the number of elements of each column of the DataFrame
```

```python
print(newdf3.max())       ### Prints the largest element of each column of the DataFrame
```

```python
print(newdf3.min())       ### Prints the smallest element of each column of the DataFrame
```

```python
print(newdf3.median())    ### Prints the Median of each column of the DataFrame
```

```python
print(newdf3.std())       ### Prints the Standard Deviation of each column of the DataFrame
```

## FreeCodeCamp Tutorial:

```python
################## Loading Data in Pandas

# poke = pd.read_csv("pokemon_data.csv")                     ### Reads .csv file

# poke = pd.read_csv("pokemon_data.txt", delimiter='\t')     ### Reads .txt file

poke = pd.read_excel("pokemon_data.xlsx")                    ### Reads Excel file

print(poke)
```

```python
poke.head()
```

```python
print(poke.columns)    ### Prints all the columns of the DataFrame
```

```python
print(poke["Name"])     ### Prints the column "Name" of the DataFrame
```

```python
print(poke.iloc[2:5])      ### Prints Rows 2,3,4 of the DataFrame
```

```python
print(poke.iloc[2,1])    ### Prints an element at Row-2 and Column-1 of the DataFrame
```

```python
poke.drop(["#"], axis = 1)    ### Drops the column "#" from the DataFrame
```

```python
print(poke)
```

```python
### Takes columns 4 to 9 and sums these elements row-wise and assigns this final value to a variable 
### and makes a new column at the end of the DataFrame

poke['Total'] = poke.iloc[:,4:10].sum(axis=1) 
print(poke)
```

```python
poke = poke.drop(["Total"], axis=1)
print(poke)
```

```python
poke['Total'] = poke.iloc[:,4:10].sum(axis=1)
print(poke)
```

```python
cols = list(poke.columns)                      ### takes all the column names in a list
poke = poke [cols[0:4]+[cols[-1]]+cols[4:12]]  ### adds last column after 3rd column
print(poke)
```

```python
### Lists all the elements with Type 1 = Grass and Type 2 = Poison 
### (and statement-executes only if both conditions are true)
### (& = and, | = or)
### (or statement-executes if any of the conditions are true)

poke.loc[(poke["Type 1"] == "Grass") & (poke["Type 2"] == "Poison")]   
```
