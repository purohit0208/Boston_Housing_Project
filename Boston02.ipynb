```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = housing.iloc[:,:-1]
print(x)
print(x.shape)
```

```python
y = housing.iloc[:,-1]
print(y)
print(y.shape)
```

```python
# plots scatter plots for 4 different attributes and shows their correlation

from pandas.plotting import scatter_matrix

attributes = ['MEDV', 'RM', 'NOX', 'LSTAT']
scatter_matrix(housing[attributes], figsize=(12,8))
plt.show()
```

```python
plt.scatter(housing['LSTAT'], housing['MEDV'])

plt.title("% lower status of the population vs Housing Price")
plt.xlabel("% lower status of the population")
plt.ylabel("Housing Price (*1000 USD)")
plt.show()
```

```python
# Uses seaborn to plot linear regression plot
# It shows RM vs Price and also linear regression line passing through data

import seaborn as sns

sns.regplot(x='CRIM', y='MEDV', data=housing)

plt.title("Crime Rate vs Housing Price")
plt.xlabel("Crime Rate")
plt.ylabel("Housing Price (*1000 USD)")
plt.show()
```

```python
## Train-Test Splitting:
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
print(x_train)
print(x_train.shape)
print(x_test.shape)
```

```python
print(y_train)
print(y_train.shape)
print(y_test.shape)
```

```python
## Standardize the dataset
## Because we use gradient discent internally, so we need to reach global minima 
## and for that we need all of our dataset variables to be of the same scale
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
print(x_train)
print(x_train.shape)
```

```python
print(x_test)
print(x_test.shape)
```

```python
print(y_train)
print(y_train.shape)
```

```python
print(y_test)
print(y_test.shape)
```

```python
## Model Training
from sklearn.linear_model import LinearRegression

regression = LinearRegression()
regression.fit(x_train, y_train)
```

```python
reg_predicted = regression.predict(x_test)         ### reg_predicted = y_predicted

print(f"reg_Predicted: {reg_predicted.shape}.")
```

```python
plt.scatter(y_test, reg_predicted)

plt.title("y_test vs reg_predicted")
plt.xlabel("y_test")
plt.ylabel("reg_predicted")
plt.show()
```

```python
print("Weights: ", regression.coef_)                   ### Prints weights (w1, w2, ..., wn)
print("Intercept: ", regression.intercept_)            ### Prints Intercept (w0)
```
